湖の窒素濃度の推移を抽出する。

データは`../水質`にある。


```python
# 相模湖
import pandas as pd

pd.set_option("display.max_rows", None)


def lake_loader(lake="10_sagamiko", loc="湖央東部"):
    dfs = pd.DataFrame()
    for year in range(21):
        with open(f"{BASEDIR}../水質/{lake}{year:02d}.xlsx", "rb") as f:
            df = pd.read_excel(f, header=1, sheet_name=loc)
            df = df.dropna(subset=["採取位置"])
            df0 = df[df["採取位置"].str.startswith("上層")]
            df1 = df0[df0["採取位置"].str.contains("下層") == False]
            df2 = df[df["採取位置"].str.startswith("下層")]
            # print(df.columns)
            if year == 21:
                df = df1[["採取年月日", "全窒素"]]
                df["下層"] = df2["全窒素"].to_numpy()
                df.columns = ["date", "上層", "下層"]
            else:
                df = df1[["測定項目", "全窒素"]]
                df["下層"] = df2["全窒素"].to_numpy()
                df.columns = ["date", "上層", "下層"]
            # if lake=="11_tsukuiko" and year==11:
            #     print(year)
            #     print(df)
            print(lake, year, len(df))
            dfs = pd.concat([dfs, df])
    return dfs


日連大橋 = lake_loader(lake="10_sagamiko", loc="日連大橋")
名手橋 = lake_loader(lake="11_tsukuiko", loc="名手橋")
相模湖 = lake_loader(lake="10_sagamiko", loc="湖央東部")
津久井湖 = lake_loader(lake="11_tsukuiko", loc="湖央部")
```

```python
# まずトレンドとそれ以外に分ける。
from statsmodels.tsa.filters.hp_filter import hpfilter
import matplotlib.pyplot as plt

temp = 日連大橋.set_index("date")
temp = temp["下層"]

cyclic, trend = hpfilter(temp)

fig, ax = plt.subplots(1, 2, figsize=(16, 4))
cyclic.plot(ax=ax[0], title="日連大橋 Cyclic Component")
trend.plot(ax=ax[1], title="日連大橋 Trend Component")
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose, STL

decomposed = seasonal_decompose(temp, model="additive", period=12)

plot = decomposed.plot()
```

年周変化が弱い。振幅が 0.1 ぐらいしかない。今のモデルではこの振動を大気窒素がもたらしていると考えるので、大気の寄与率がとても小さいことになってしまう。
