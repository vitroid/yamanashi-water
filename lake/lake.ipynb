湖の窒素濃度の推移を抽出する。

データは`../水質`にある。

```python
# 相模湖
import pandas as pd

pd.set_option("display.max_rows", None)
loc = "湖央東部"
dfs = pd.DataFrame()
for year in range(21):
    print(year)
    with open(f"{BASEDIR}../水質/10_sagamiko{year:02d}.xlsx", "rb") as f:
        df = pd.read_excel(f, header=1, sheet_name=loc)
        df = df.dropna(subset=["採取位置"])
        df0 = df[df["採取位置"].str.startswith("上層")]
        df1 = df0[df0["採取位置"].str.contains("下層") ==False]
        df2 = df[df["採取位置"].str.startswith("下層")]
        # print(df.columns)
        if year == 21:
            df = df1[["採取年月日", "硝酸性窒素及び亜硝酸性窒素"]]
            df["下層"] = df2["硝酸性窒素及び亜硝酸性窒素"].to_numpy()
            df.columns = ["date", "上層", "下層"]
        else:
            df = df1[["測定項目", "亜硝酸性窒素及び硝酸性窒素"]]
            df["下層"] = df2["亜硝酸性窒素及び硝酸性窒素"].to_numpy()
            df.columns = ["date", "上層", "下層"]
        # print(year)
        # print(df)
        # print(df)            
        dfs = pd.concat([dfs, df])
dfs
```

```python
flows = pd.read_csv(f"{BASEDIR}flow/9bridges2000.csv") # 2000年4月以降の252ヵ月

bridges = ["桂川橋", "道志川", "秋山川"]

Fs = None
Ns = None
for bridge in bridges:
    flowcol = f"('{bridge}', '流量')"  # 単位はm3/s
    noxcol = f"('{bridge}', '全窒素')"  # 単位はppm ~ g/m3

    # River flow
    F = flows[flowcol]
    if Fs is None:
        Fs = F
    else:
        Fs += F

    # N load
    N = flows[flowcol] * flows[noxcol]
    if Ns is None:
        Ns = N
    else:
        Ns += N

# river N conc.
cr = Ns/Fs

cr
```
